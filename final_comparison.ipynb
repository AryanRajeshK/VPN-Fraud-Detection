```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

models = ["GBT", "Random Forest", "Decision Trees", "KNN", "CNN", 
          "SVM", "RNN", "MLP", "Logistic Regression", "Naive Bayes"]
accuracies = [0.9261, 0.9247, 0.8939, 0.8305, 0.8008, 0.7814, 0.7434, 0.7298, 0.6411, 0.5423]
precisions = [0.93,   0.92,   0.89,   0.83,   0.81,   0.79,   0.75,   0.73,   0.65,   0.67]
recalls    = [0.93,   0.92,   0.89,   0.83,   0.80,   0.78,   0.74,   0.73,   0.64,   0.54]
f1_scores  = [0.93,   0.92,   0.89,   0.83,   0.80,   0.78,   0.74,   0.73,   0.63,   0.41]
accuracies_pct = [a * 100 for a in accuracies]
precisions_pct = [p * 100 for p in precisions]
recalls_pct    = [r * 100 for r in recalls]
f1_scores_pct  = [f * 100 for f in f1_scores]

group_positions = np.arange(len(models))
width = 0.15  
fig, ax = plt.subplots(figsize=(14, 7))
ax.set_axisbelow(True)  

colors = {
    'Accuracy': '#5A5A5A',
    'Precision': '#4F81BD',
    'Recall': '#C0504E',
    'F1 Score': '#9BBB59'
}

# Plot each metric as a separate set of bars
ax.bar(group_positions - 1.5 * width, accuracies_pct, width, label='Accuracy', color=colors['Accuracy'], zorder=3)
ax.bar(group_positions - 0.5 * width, precisions_pct, width, label='Precision', color=colors['Precision'], zorder=3)
ax.bar(group_positions + 0.5 * width, recalls_pct, width, label='Recall', color=colors['Recall'], zorder=3)
ax.bar(group_positions + 1.5 * width, f1_scores_pct, width, label='F1-Score', color=colors['F1 Score'], zorder=3)
ax.set_ylabel('Percentage (%)', fontsize=12)
ax.set_title('Performance Comparison', fontsize=14)
ax.set_xticks(group_positions)
ax.set_xticklabels(models, fontsize=11, rotation=45)
ax.legend(fontsize=11)
ax.yaxis.set_major_locator(MultipleLocator(10))
ax.grid(axis='y', linestyle='--', color='gray', alpha=0.7, zorder=0)

fig.tight_layout()
plt.show()
```
